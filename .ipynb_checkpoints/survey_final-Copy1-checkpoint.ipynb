```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.patches as mpatches
import seaborn as sns
```

```python
# Create path to .csv file
path = "data/mental_health_in_tech.csv"

# Read data
mental_health_data = pd.read_csv(path)
mental_health_data
```

```python
# Begin data cleanup
mental_health_data.drop(columns=["Timestamp", "Age", "Gender", "Country", "no_employees", "care_options",
 "wellness_program", "mental_health_consequence", "phys_health_consequence", "coworkers", "supervisor",
 "mental_health_interview","phys_health_interview","mental_vs_physical","obs_consequence","seek_help", "comments", "self_employed", "leave", "work_interfere"], axis=1, inplace=True)

mental_health_data
```

```python
# Restrict the data to 2014 only 
data_year = mental_health_data.loc[mental_health_data['year']==2014]
data_year
```

```python
# Restrict the data to August and September only
data_month = data_year[data_year.month.isin(['Aug','Sep'])]
data_month_df=pd.DataFrame(data_month)
data_month_df
```

```python
# Restrict age data 18-72
# data_age = df[df.Age.between(18, 25)]
data_age = data_month_df.dropna(subset=['age'], how='any')
data_age_df = pd.DataFrame(data_age)
data_age_df
```

```python
# Get a list of columns 
cols = list(data_age_df.columns.values)
cols
```

```python
# Rearrange columns 
data_cleaned_df = data_age_df[['gender','age','state','month','year','family_history',
                              'treatment','age_band','remote_work','tech_company','benefits',
                              'anonymity','company_size','country_rollup']]
data_cleaned_df
```

```python
# Gender statistics
gender_distribution = data_cleaned_df['gender'].value_counts()
gender_distribution
```

```python
# Plot gender distribution 
colors = ['#b760bf','#d498d9','#f0d0f2']
explode = (0.05,0.05,0.05)
gender_pie = gender_distribution.plot.pie(y=gender_distribution, autopct="%1.1f%%", colors = colors, title = 'Mental Health Data Gender Distribution', shadow=False, startangle=45)

# Get the current figure
gender_pie = plt.gcf()

# Draw a circle so it looks nicer 
centre_circle = plt.Circle((0,0),0.75,fc='white')

# Add the circle
gender_pie.gca().add_artist(centre_circle)

# Increase circle size 
gender_pie.set_size_inches(7,7)

# Remove y-axis label and change layout
plt.ylabel("")

# Ajust subplot parameters so it fits in the figure 
plt.tight_layout()

# Save figure
plt.savefig("output_data/gender_distribution.png")

plt.show()
```

```python
# Get the total treatment percentage 
total_treatment = data_cleaned_df['treatment'].value_counts(normalize=True) * 100
total_treatment.round(2).map("{:,.2f}%".format)
```

```python
# Generate a chart for total percentage 
color = '#d178d6'
total_treatment_chart = total_treatment.plot(kind='bar', legend=False, 
                                            rot="horizontal",
                                            title = "Have you sought treatment for a mental condition?",
                                            ylim=(0,60), color=color)


# Add y-axis label and title
total_treatment_chart.set_ylabel('Response Percentage')
total_treatment_chart.set_xlabel('Obtained Treatment')

# Get the current figure
total_treatment_chart = plt.gcf()

# Adjust size of figure
total_treatment_chart.set_size_inches(7,7)

# Ajust subplot parameters so it fits in the figure 
plt.tight_layout()

# Save figure 
plt.savefig("output_data/total_treatment_distribution.png")

plt.show()
```

```python
# Groupby gender 
gender_group = data_cleaned_df.groupby('gender')

# Get the count of yes/no per gender 
treatment_count = gender_group['treatment'].value_counts()

# Create new DataFrame with the counts 
gender_group_treatment_df = pd.DataFrame({"Treatment Count":treatment_count})
gender_group_treatment_df
```

```python
# Generating bar chart for treatments 
color = '#d178d6'
gender_group_treatment_chart = gender_group_treatment_df.plot(kind='bar',rot=90, legend=False, color = color,
                                                              title = "Have you sought treatment for a mental condition?")


# Add y-axis label and title
gender_group_treatment_chart.set_ylabel('Response Count')
gender_group_treatment_chart.set_xlabel('Obtained Treatment')

# Get the current figure
gender_group_treatment_chart = plt.gcf()

# Adjust size of figure
gender_group_treatment_chart.set_size_inches(7,7)

# Ajust subplot parameters so it fits in the figure 
plt.tight_layout()

# Save figure 
plt.savefig("output_data/treatment_distribution.png")

plt.show()
```

```python
# Get percentages for the treatment distribution
treatment_count_percentages = gender_group['treatment'].value_counts(normalize=True) * 100

# treatment_count_percentages = treatment_count_percentages.map("{:,.2f}%".format)

# Create new DataFrame with the counts 
treatment_percentage_df = pd.DataFrame({"Treatment Percentage":treatment_count_percentages}).round(2)
treatment_percentage_df
```

```python
# Generating bar chart with percentages for treatments 
color = '#d178d6'
treatment_percentage_chart = treatment_percentage_df.plot(kind='bar',rot=90, legend=False, color = color,
                                                         title = "Have you sought treatment for a mental condition?",
                                                         ylim=(0,100))

# Add y-axis label and title
treatment_percentage_chart.set_ylabel('Response %')
treatment_percentage_chart.set_xlabel('Obtained Treatment')

# Get the current figure
treatment_percentage_chart = plt.gcf()

# Adjust size of figure
treatment_percentage_chart.set_size_inches(7,7)

# Ajust subplot parameters so it fits in the figure 
plt.tight_layout()

# Save figure 
plt.savefig("output_data/treatment_distribution_percentage.png")

plt.show()
```

```python
# Grouped gender/tech/treatment 
group_gtt = data_cleaned_df.groupby(['gender','treatment']).size()

# Create dataframe 
group_gtt_df = pd.DataFrame({'Results':group_gtt})
group_gtt_df
```

```python
# Create a stacked bar chart for hypothesis 1
colors = ['#b760bf', '#d498d9']
group_gtt_chart = group_gtt_df.unstack().plot.bar(stacked=True, legend = False,color = colors, figsize=(8,8))

# Label
group_gtt_chart.set_ylabel('Total Number of Responses')
group_gtt_chart.set_xlabel('Treatment Obtained by Gender')
group_gtt_chart.set_title('Mental Health in Tech - Responses by Gender')

# Save figure 
plt.savefig("output_data/seeking_out_mental_health.png")

plt.show()
```
